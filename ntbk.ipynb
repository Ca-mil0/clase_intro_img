cargar imagen con open cv

```python
x=cv2.imread("data/unalapa.png")
```

```python
x.shape
```

```python
#convertir a bgr a rgb

x_rgb = cv2.cvtColor(x, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(5,5))
plt.subplot(1,2,1)
plt.imshow(x)
plt.axis('off')
plt.title('BGR')
plt.subplot(1,2,2)
plt.imshow(x_rgb)
plt.axis('off') 
plt.title('RGB')
```

```python
x_crop_ver=x_rgb[:, 0:512, :]
x_crop_ver.shape
```

```python
plt.figure(figsize=(5,5))
plt.imshow(x_crop_ver)
plt.axis('off')
```

```python
x_crop_vert=x_rgb[:, 512:1024, :]
x_crop_vert.shape
plt.figure(figsize=(5,5))
plt.imshow(x_crop_vert)
plt.axis('off')
```

```python
plt.figure(figsize=(5,5))
plt.imshow(x_crop_ver)
plt.imshow(x_crop_vert, alpha=0.5)
plt.axis('off')
plt.show()
```

```python
xconcat=np.concatenate((x_crop_ver, x_crop_vert), axis=1)
plt.figure(figsize=(5,5))
plt.imshow(xconcat)
plt.title("imagenes pegadas")
plt.axis("off")
```

```python
x_blue=x_rgb.copy()
x_blue[:,:,0]=0
x_blue[:,:,1]=0
plt.figure(figsize=(5,5))
plt.imshow(x_blue)
plt.title("canal azul")
plt.axis("on")
```

```python
x_rd=x_rgb.copy()
x_rd[:,:,2]=0
x_rd[:,:,1]=0
plt.figure(figsize=(5,5))
plt.imshow(x_rd)
plt.title("canal rojo")
plt.axis("on")
```

```python
x_gn=x_rgb.copy()
x_gn[:,:,0]=0
x_gn[:,:,2]=0
plt.figure(figsize=(5,5))
plt.imshow(x_gn)
plt.title("canal verde")
plt.axis("on")
```
